```python
fig, ax = plt.subplots(1,2, figsize = (12,6))
for i in set(df.Country):
    df_country = df[df.Country == i]
    if i != 'Zimbabwe':
        ax[0].plot(df_country.Year, df_country['Life expectancy at birth (years)'], label = i, marker = 'o')
        ax[0].legend()
    else:
        ax[1].plot(df_country.Year, df_country['Life expectancy at birth (years)'], label = i, marker = 'o')
        ax[1].legend(loc = 'lower right')
plt.title('Life Expectancy')
```
